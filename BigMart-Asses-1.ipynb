```python
non_numerical_features = data.select_dtypes(include=[np.object])
non_numerical_features.head()
numerical_features = data.select_dtypes(include=[np.float64])
numerical_features.corr()
```

```python
for col in non_numerical_features:
    dummy = pd.get_dummies(data[col])
    data = data.drop(col,axis=1)
    data = data.join(dummy)
```

```python
y = data.Item_Outlet_Sales
x = data.drop('Item_Outlet_Sales',axis =1)
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state = 3)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn import metrics
```

```python
lr = LinearRegression(normalize = True)
lr.fit(x_train,y_train)
lr_y_prediction = lr.predict(x_test)

print("Linear Regression ")
print('R- square value ' ,metrics.r2_score(y_test,lr_y_prediction))
print('Mean Square Error ' ,metrics.mean_squared_error(y_test,lr_y_prediction))
print('\n')


# Tried for a lot of alpha values, but found 0.05 as the best alpha value.

alpha_list = [0.005]
for a in alpha_list:

    
    la = Lasso(normalize=True,alpha = a,max_iter=10000)
    rr = Ridge(normalize=True,alpha = a)

    
    la.fit(x_train,y_train)
    rr.fit(x_train,y_train)



    
    la_y_prediction = la.predict(x_test)
    rr_y_prediction = rr.predict(x_test)

    
   
    print("Lasso Regression for alpha = ",a)
    print('R- square value ' ,metrics.r2_score(y_test,la_y_prediction))
    print('Mean Square Error ' ,metrics.mean_squared_error(y_test,la_y_prediction))
    print('\n')
    print("Ridge Regression for alpha = ",a)
    print('R- square value ' ,metrics.r2_score(y_test,rr_y_prediction))
    print('Mean Square Error ' ,metrics.mean_squared_error(y_test,rr_y_prediction))
    print('\n')
    print('-'*40)
```

```python

# What have I plotted ?

plt.subplot(2,2,1)
plt.plot(x_test,lr_y_prediction,color = 'red')
plt.title("Linear Regression")
plt.show()


plt.subplot(2,2,2)
plt.plot(x_test,la_y_prediction,color = 'red')
plt.title("Lasso Regression")
plt.show()

plt.subplot(2,2,4)
plt.plot(x_test, rr_y_prediction,color = 'red')
plt.title("Ridge Regression")
plt.show()
```
